```python
import cv2
import matplotlib.pyplot as plt

# Load low-resolution image
image = cv2.imread('/content/Malayalam Paper.jpg')

# Upscale using different interpolation techniques
upscaled_image_linear = cv2.resize(image, (0, 0), fx=2, fy=2, interpolation=cv2.INTER_LINEAR)
upscaled_image_cubic = cv2.resize(image, (0, 0), fx=2, fy=2, interpolation=cv2.INTER_CUBIC)
upscaled_image_lanczos = cv2.resize(image, (0, 0), fx=2, fy=2, interpolation=cv2.INTER_LANCZOS4)

# Display the images
plt.figure(figsize=(10, 8))
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(upscaled_image_linear, cv2.COLOR_BGR2RGB))
plt.title('Linear Interpolation')
plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(upscaled_image_cubic, cv2.COLOR_BGR2RGB))
plt.title('Cubic Interpolation')
plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(upscaled_image_lanczos, cv2.COLOR_BGR2RGB))
plt.title('Lanczos Interpolation')
plt.show()
```

```python
import cv2
from google.colab.patches import cv2_imshow

# Load image with noise
image = cv2.imread('/content/Malayalam Paper.jpg')

# Denoise the image using OpenCV's fastNlMeansDenoising function
denoised_image = cv2.fastNlMeansDenoisingColored(image, None, 10, 10, 7, 21)

# Save and display the denoised image
cv2.imwrite('denoised_image.jpg', denoised_image)
cv2_imshow(denoised_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import tensorflow as tf
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load a pre-trained ESRGAN model
model = tf.keras.models.load_model('path_to_pretrained_ESRGAN_model.h5')

# Load low-resolution image
image = cv2.imread('/content/Malayalam Paper.jpg')

# Convert to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Preprocess the image
image_resized = cv2.resize(image_rgb, (image_rgb.shape[1] * 2, image_rgb.shape[0] * 2))  # Double the resolution
image_resized = np.expand_dims(image_resized, axis=0)  # Add batch dimension

# Perform super-resolution
upscaled_image = model.predict(image_resized)

# Convert back to uint8 for display
upscaled_image = np.squeeze(upscaled_image, axis=0).astype(np.uint8)

# Display the original and upscaled image
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.subplot(1, 2, 2)
plt.imshow(upscaled_image)
plt.title('Upscaled Image using ESRGAN')
plt.show()
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Load image
image = cv2.imread('/content/Malayalam Paper.jpg', cv2.IMREAD_GRAYSCALE)

# Apply contrast stretching
min_value = np.min(image)
max_value = np.max(image)

# Stretching formula
contrast_stretched = np.uint8((image - min_value) * 255 / (max_value - min_value))

# Display the result
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.subplot(1, 2, 2)
plt.imshow(contrast_stretched, cmap='gray')
plt.title('Contrast Stretched Image')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Load image in grayscale
image = cv2.imread('/content/Malayalam Paper.jpg', cv2.IMREAD_GRAYSCALE)

# Apply Histogram Equalization
equalized_image = cv2.equalizeHist(image)

# Display the images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.subplot(1, 2, 2)
plt.imshow(equalized_image, cmap='gray')
plt.title('Histogram Equalized Image')
plt.show()
```

```python
import tensorflow as tf
import cv2
import matplotlib.pyplot as plt

# Load a pre-trained GAN model (Pix2Pix or CycleGAN)
generator = tf.keras.models.load_model('path_to_pretrained_GAN_model.h5')

# Load low-resolution image
image = cv2.imread('low_resolution_image.jpg')

# Preprocess the image
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_resized = cv2.resize(image_rgb, (256, 256))  # Resize to model input size
image_resized = np.expand_dims(image_resized, axis=0)  # Add batch dimension

# Perform enhancement using the GAN model
enhanced_image = generator.predict(image_resized)

# Convert back to uint8 for display
enhanced_image = np.squeeze(enhanced_image, axis=0).astype(np.uint8)

# Display the result
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.subplot(1, 2, 2)
plt.imshow(enhanced_image)
plt.title('Enhanced Image using GAN')
plt.show()
```
