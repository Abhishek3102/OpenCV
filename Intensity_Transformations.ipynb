```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt

# Read the image
image = cv2.imread('/content/sesko.jpeg', cv2.IMREAD_GRAYSCALE)
```

```python
# Display original image
plt.figure(figsize=(10,10))
plt.subplot(2, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')
```

```python
# Display original image
plt.figure(figsize=(10,10))
plt.subplot(2, 2, 1)
plt.imshow(image)
plt.title('Original Image')
plt.axis('off')
```

```python
# 1. Image Negative (Inversion)
negative_image = 255 - image
plt.subplot(2, 2, 2)
plt.imshow(negative_image, cmap='gray')
plt.title('Image Negative')
plt.axis('off')
```

```python
# 2. Contrast Stretching
min_pixel = np.min(image)
max_pixel = np.max(image)

# Define a scaling function for contrast stretching
contrast_stretched = ((image - min_pixel) / (max_pixel - min_pixel) * 255).astype(np.uint8)
plt.subplot(2, 2, 3)
plt.imshow(contrast_stretched, cmap='gray')
plt.title('Contrast Stretching')
plt.axis('off')
```

```python
# 3. Thresholding
_, thresholded_image = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)
plt.subplot(2, 2, 4)
plt.imshow(thresholded_image, cmap='gray')
plt.title('Thresholding')
plt.axis('off')
```

```python
# 4. Gray-Level Slicing
# Only the pixels in a given range will be kept, others will be black
lower_bound = 100
upper_bound = 150
sliced_image = np.where((image >= lower_bound) & (image <= upper_bound), 255, 0).astype(np.uint8)

plt.figure(figsize=(10,10))
plt.imshow(sliced_image, cmap='gray')
plt.title('Gray-Level Slicing (100-150)')
plt.axis('off')

# Show all images
plt.show()
```
